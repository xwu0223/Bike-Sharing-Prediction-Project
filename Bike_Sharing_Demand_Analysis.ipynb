```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
x1,y1 = ecdf(hour.temp)
plt.plot(x1,y1,marker='.',linestyle='none')
plt.xlabel('temp')
plt.ylabel('ECDF')
```

```python
x1,y1 = ecdf(hour.hum)
plt.plot(x1,y1,marker='.',linestyle='none')
plt.xlabel('humidity')
plt.ylabel('ECDF')
```
